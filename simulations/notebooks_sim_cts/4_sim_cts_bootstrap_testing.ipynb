## Two sample test based on Bootstrap

```r
## interpretation: if confidence interval doesn't contain zero, then significant
## ref: correct intrepretation of confidence intervals: https://journals.sagepub.com/doi/pdf/10.1177/201010581001900316
```

```r
diff_toe = (table$toe_lin - table$toe_rf)
mean(diff_toe)
quantile(diff_toe, probs = c(0.025, 0.975))
```

```r
diff_block = (table$block_lin - table$block_rf)
mean(diff_block)
quantile(diff_block, probs = c(0.025, 0.975))
```
